Are there more dogs surrendered to the shelter during certain times of the year? Specificially looking at Christmas... after dogs are given as presents, does shelter intake increase?

```python
import numpy as np
import pandas as pd
from pandas import DataFrame
import seaborn as sns
```

```python
shelterdframe = pd.read_csv('aac_intakes_outcomes.csv')
```

```python
shelterdframe.head()
```

```python
dogs_only_dframe = shelterdframe.loc[shelterdframe['animal_type'] == 'Dog']
```

```python
dogs_only_dframe
```

```python
sns.countplot('intake_month',data=dogs_only_dframe)
```

```python
months = {1:'January',
          2:'February',
          3:'March',
          4:'April',
         5:'May',
         6:'June',
         7:'July',
         8:'August',
         9:'September',
         10:'October',
         11:'November',
         12:'December'}

dogs_only_dframe['Month']= dogs_only_dframe['intake_month'].map(months)
```

```python
dogs_only_dframe.head()
```

```python
intake_by_month = sns.countplot('Month',
                                data=dogs_only_dframe,
                                order=['January','February','March','April','May','June','July','August','September','October','November','December'])

intake_by_month.set_xticklabels(intake_by_month.get_xticklabels(), rotation=40, ha="right")
```

In order to find out more about the time of year more dogs are surrendered, I first filtered the dataframe to find entries for dogs only. Based on the provided intake month, I was able to make a simple bar graph showing the intake count for each month. I was expecting to see the highest intake count in January, directly following Christmas. While the count is one of the highest, there are similar amounts in other times of the year. While I am unable to draw any conclusions regarding my Christmas question, the data does show an interesting trend of lower values between April and September.

```python
dogs_and_cats = shelterdframe[(shelterdframe.animal_type.isin(["Dog", "Cat"]))]
```

```python
months = {1:'January',
          2:'February',
          3:'March',
          4:'April',
         5:'May',
         6:'June',
         7:'July',
         8:'August',
         9:'September',
         10:'October',
         11:'November',
         12:'December'}

dogs_and_cats['Month']= dogs_and_cats['intake_month'].map(months)
```

```python
combined_intake = sns.countplot('Month',
                                data=dogs_and_cats,
                                order=['January','February','March','April','May','June','July','August','September','October','November','December'])

combined_intake.set_xticklabels(combined_intake.get_xticklabels(), rotation=40, ha="right")
```
